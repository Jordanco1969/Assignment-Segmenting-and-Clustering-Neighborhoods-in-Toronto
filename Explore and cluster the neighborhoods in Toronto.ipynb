TORONTO LATITUDE AND LONGITUDE

```python
Latitude, Longitude =43.651070, -79.347015

map_Toronto = folium.inline_maps(location=[Latitude, Longitude], zoom_start=10)
```

add markers to map

```python
for pos_code, borough, neighborhood, lat, lng  in df.values:
    
#    print(pos_code, borough, neighborhood, lat, lng)
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
#print(map_Toronto)
map_Toronto
```
